```python

import pandas as pd
data = pd.read_csv('Nutrition__Physical_Activity__and_Obesity_-_Behavioral_Risk_Factor_Surveillance_System.csv')
data.head()
                    
```

```python
data.info()
```

```python
data['Question'].unique()
```

```python
data.drop(['YearEnd','Total','LocationID','LocationAbbr','Datasource','Topic','Data_Value_Alt','Data_Value_Unit','QuestionID','ClassID','TopicID','GeoLocation','DataValueTypeID','Data_Value_Type','Data_Value_Footnote_Symbol','Data_Value_Footnote','StratificationCategoryId1','StratificationID1'],axis =1,inplace = True);
```

```python
data[data['Education'].isna() == False]
```

```python
#predictions based on gender
genderGroup = data[(data['Class'] =='Obesity / Weight Status') & (data['StratificationCategory1']=='Gender')]
```

```python
genderGroup = genderGroup.groupby(['YearStart','Stratification1'])['Data_Value'].agg(mean='mean')
genderGroup.reset_index(inplace=True)
```

```python
genderGroup
```

```python
import numpy as np
X_data = genderGroup.copy()
X_data["Stratification1"] = np.where(X_data["Stratification1"] == "Female", 0, 1)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_data.iloc[:,0:2],X_data.iloc[:,2])
```

```python
lr.predict([[2030,0],[2030,1]])
```

```python
#predictions based on age range over years
ageGroup = data[(data['Class'] =='Obesity / Weight Status') & (data['StratificationCategory1']=='Age (years)')]
ageGroup = ageGroup.groupby(['YearStart','Age(years)'])['Data_Value'].agg(mean='mean')
ageGroup.reset_index(inplace=True)
ageGroup
```

```python
from sklearn.preprocessing import LabelEncoder
le_ageGroup = LabelEncoder()
ageGroup['Age(years)'] = le_ageGroup.fit_transform(ageGroup['Age(years)'])
```

```python
lr = LinearRegression()
lr.fit(ageGroup.iloc[:,0:2],ageGroup.iloc[:,2])
```

```python
lr.predict([[2030,4]])
```
